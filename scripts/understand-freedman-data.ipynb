```python
# need to confirm that the runs match the samples

# first, download the series matrix file : https://ftp.ncbi.nlm.nih.gov/geo/series/GSE130nnn/GSE130408/matrix/GSE130408_series_matrix.txt.gz
# read it in 

series_data = pd.read_table('./GSE130408_series_matrix.txt.gz', skiprows=28, header=0)
series_data.head(10)
```

```python
series_data.loc[series_data['!Sample_title'] == '!Sample_relation'] # I need the second one since there are 2 of them
```

```python
series_sra = [sra.split('=')[1] for sra in series_data.loc[series_data['!Sample_title'] == '!Sample_relation'].iloc[1, :].astype('str').values[1:]]
series_sra
```

```python
hw = subprocess.run(f'ls ./experiment_details/ | wc -l', shell=True, capture_output=True, text=True)
int(hw.stdout.split('\n')[0])
```

```python
#sras = series_sra #['SRX8406457', 'รง']
# make the experiment_details directory

# there should be 249 json files in the directory
# if not, run the entire thing
hw = subprocess.run(f'ls ./experiment_details/ | wc -l', shell=True, capture_output=True, text=True)
if int(hw.stdout.split('\n')[0]) == 249:
    print(f'There are {249} json files in the directory.')
else:
    print('There are {} files instead of 249 files.')
    for sra in series_sra:
        if not os.path.isfile(f'./experiment_details/{sra}.json'):
            subprocess.run(f'ffq {sra} > ./experiment_details/{sra}.json', shell=True)
```

```python
# read the files individually
json_files = os.listdir('./experiment_details/')
SRR_nums = []
for sra in series_sra:
    json_file = open(f'./experiment_details/{sra}.json')
    if os.stat(f'./experiment_details/{sra}.json').st_size == 0:
        SRR_nums.append('ERROR')
        continue
    json_data = json.load(json_file)
    SRR_nums.append(list(json_data[f'{sra}']['runs'].keys())[0])
    json_file.close()
SRR_nums
```

```python
len(SRR_nums)
```

```python
# for some reason, the SRR for these files could not be gotten, here are their indices
empty_files = [i for i, S in enumerate(SRR_nums) if S == 'ERROR']
empty_files
```

```python
# for now..
GSM_num = series_data.iloc[0, : ]
GSM_num.name = 'GSM'
GSM_num = pd.DataFrame(GSM_num).iloc[1:, :]
```

```python
SRR_num = pd.DataFrame(SRR_nums, columns=['SRR'], index=list(GSM_num.index))
```

```python
exp_details = pd.concat([GSM_num, SRR_num], axis=1)
exp_details.iloc[empty_files, :]
```
