```python
import pandas as pd, re
```

```python
GATA3_motifs = pd.read_table('/projects/covid-ct/imlab/users/temi/projects/TFXcan/processed-data/Gata3_motif_regions_TEMP.txt', 
                             sep=' ', names=['chr', 'start', 'end', 'id', 'score', 'strand', 'motif_name', 'pos'])
GATA3_motifs
```

```python
def create_bcftools_input(motif_regions=GATA3_motifs, sample_name=None, output_dir='/projects/covid-ct/imlab/users/temi/projects/TFXcan/code_arguments'):
    '''
    Creates the input for creating the per individual-chromosome vcf files
    
    Return:
        Nothing. File is written in the directory. 
    '''
    if isinstance(None, type(sample_name)):
        sample_name = 'Not_supplied'
        
    a = pd.Series([re.sub(pattern='^[chr]*', repl='', string=G) for G in motif_regions.chr.values])
    b = motif_regions.start.astype(str)
    c = motif_regions.end.astype(str)
    
    d = a + ':' + b + '-' + c
    
    with open(f'{output_dir}/{sample_name}_bed.txt', 'w') as txt:
        txt.write(sample_name + ' ' + ' '.join(d.to_list()))
        
    print('Done writing.')
```

```python
create_bcftools_input(sample_name='HG00096')
```
